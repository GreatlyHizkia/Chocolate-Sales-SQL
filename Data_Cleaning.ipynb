```python
import pandas as pd

# Read File
df = pd.read_csv("/content/Chocolate Sales.csv")

# Clear the ‘Amount’ column of ‘$’ and ‘,’ then convert to float
df['Amount'] = (
    df['Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
    .astype(float)
)
# Save DataFrame to CSV
df.to_csv("Chocolate_Sales_Cleaned.csv", index=False)
```
